# Handle Null Values

```python
# check for missing values
null = df.isna().sum()
null_percentage = round(null*100/len(df),2)
null_df = pd.DataFrame({'Null Values':null, 'Percentage Null Values':null_percentage})

# display only the columns that have null values
null_df[null_df['Null Values']>0]
```

There are some missing values in our dataset, let's explore these null entries.

```python
# children
df['children'].value_counts()
```

Since there are only 4 bookings where the `children` values are missing, we can impute these missing values as the mode 0. Notice that there is one booking with 10 children, this could potentially be an outlier, we will investigate this later.

```python
df[df['children'].isna()]
```

```python
# impute missing  children
df['children'].fillna(0.0,inplace=True)
```

```python
df['children'].value_counts()
```

```python
# country
df['country'].unique()
```

Take a look at the distribution of countries.

```python
# make a dataframe of the percentage of bookings by `country`
country_counts = df['country'].value_counts()
countries = country_counts.index
percentage_bookings = country_counts.values*100/len(df)

# put in a dataframe
country_percentage = pd.DataFrame({'percentage_of_bookings':percentage_bookings.round(2)},index=countries).sort_values('percentage_of_bookings',ascending=False).head(20)
country_percentage
```

```python
# visualize 
plt.figure(figsize=(15,8))
sns.barplot(x=country_percentage.index,y=country_percentage['percentage_of_bookings'],palette='flare')
plt.title('Top 20 Countries with Most Bookings',weight='bold')
plt.xlabel('Countries')
plt.ylabel('Percentage of Total Bookings')
plt.grid(False)
plt.show()
```

Most of bookings are made by the guests from Portugual, we can impute the missing countries to the mode which is 'PRT'.

```python
# fill missing `country` with 'PRT'
df['country'].fillna('PRT',inplace=True)

# check
df['country'].isna().sum()
```

Now, `agent` column is missing roughly 14% of the data. Let's explore and find the best way to treat them.

```python
# agent
df['agent'].unique()         
```

In our scenario, it is indicative that the missing values in `agent` mean the corresponding bookings were not made by any agent. 

```python
# group canceled and non-canceled bookings by `agent`
agent = df.groupby(['agent','is_canceled'])['agent'].count().unstack().sort_values(1,ascending=False).head(15)
agent
```

```python
plt.figure(figsize=(15,8))
agent.plot(kind='barh',stacked=True,color={1:'lightcoral',0:'purple'})
plt.title('Cancellations vs Agent')
plt.xlabel('Agent')
plt.ylabel('Number of Bookings')
plt.grid(False)
plt.show()
```

Albeit cancellation rates vary among different agents. Generally speaking, poular agents have higher cancellation rate. The number of unique values in this column is quite large, which would generate a excessively complex model. However, this column still contain useful information for predicting the cancellation, i.e. bookings made through certain agents have higher cancellation rate. We should keep the feature as it is for now, and replace missing agents with 0.0. Note that the data type is in float, should be converted to integer later.

```python
# replace missing `agent` with 0.0
df['agent'] = df['agent'].apply(lambda x: 0.0 if pd.isna(x) else x)

# check
df['agent'].isna().sum()
```

```python
# company 
df['company'].unique()
```

```python
# group canceled and non-canceled bookings by `company`
company = df.groupby(['company','is_canceled'])['company'].count().unstack().sort_values(1,ascending=False).head(10)
company
```

```python
# visualize
plt.figure(figsize=(15,8))
company.plot(kind='barh',stacked=True,color={1:'lightcoral',0:'purple'})
plt.title('Cancellations vs Company')
plt.xlabel('Company')
plt.ylabel('Number of Bookings')
plt.grid(False)
plt.show()
```

For the same reason as I did to `agent` column, `company` has a lot of unique values as well, but they also contain information of our target variable, the cancellation rates vary among all the companies. But there 94% missing values in this column, this is a lot. Let's explore the feature and decide what to do with them.

```python
# check if there is 0.0 in `company`
df[df['company']==0.0]
```

Looks like all the bookings without any missing value in the `company` column were paid by a company. Now divide the data into those missing `company` data and those don't.

```python
# cancellations for data with missing values in `company` and those without
grouped_company = ((df.groupby([df['company'].isna(),'is_canceled'])['is_canceled'].count())/df.groupby(df['company'].isna())['is_canceled'].count()*100).unstack()
grouped_company
```

```python
# visulize the cancellations between two groups
plt.figure()
grouped_company.plot(kind='bar',stacked=True,color={1:'lightcoral',0:'purple'})
plt.title('Cancellation Rate in Relation with Nulled Company Column')
plt.xlabel('Missing `company`')
plt.show()
```

Looks like bookings that were paid by a company are less likely to be canceled by guests. This actually makes sense, since the trips paid by compaines are most likely to be business trips, thus the guests are more certain to show up. Since the majority of the values in `company` column are missing, and I am provided with no information of how to deal with these null values. It is better to binarize the column instead of filling in value 0.0. This way, there is still some information captured after binarization like what above visualization showed, also the complexity of the models will be reduced later on.

```python
# binarize `company`
df['company'] = df['company'].apply(lambda x: 0 if pd.isna(x) else 1)

# check
df['company'].value_counts()
```

```python
# check for missing values again
df.isna().sum()
```

## Adults

 The minimum number of `adults` is 0, taht means the bookings are made without any adult. This could be problematic.

```python
# drop bookings with no adults
df = df.drop(df[df['adults']<1].index)

# check
df['adults'].describe()
```

There is only one row with negative adr, so it is safe to drop this row

```python
# drop adr<0
df = df.drop(df[df['adr']<0].index)

# check
df['adr'].describe()
```

## Arrival Date and Reservation Status Date

Make sure the date when the guests arrived at hotel plus the total days they stayed in the hotel equals to the date when `reservation_status` changes to 'Check-Out'.

```python
# make sure all bookings that are not canceled have status 'Check-Out'
df[df['reservation_status']=='Check-Out']['is_canceled'].value_counts()
```

```python
df['arrival_date_month'].unique()
```

```python
# convert the month names to numeric month
month = {'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12,
       'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6}
df['arrival_date_month'] = df['arrival_date_month'].map(month)
```

```python
df.head()
```

```python
# combine `arrival_date_year`, `arrival_date_month`, `arrival_date_day_of_month` into one datetime column
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str)+'-'+df['arrival_date_month'].astype(str)+'-'+df['arrival_date_day_of_month'].astype(str))
```

```python
# create a column for the number of days between `arrival_date` and `reservation_status_date`
df['days_diff'] = df['reservation_status_date']-df['arrival_date']
df['days_diff'] = df['days_diff'].astype(str)
split = df['days_diff'].str.split(' ')
df['days_diff'] = split.str[0].astype(int)
```

```python
# since we are not provided with different pricing for weekdays and weekends, `stays_in_week_nights` and `stays_in_weekend_nights` can be combined into one column
df['total_stay_days'] = df['stays_in_week_nights']+df['stays_in_weekend_nights']
```

```python
df[df['total_stay_days']!=df['days_diff']]
```

Some rows have negative values in `days_diff`, let's take a look.

```python
df[df['days_diff']<0]['is_canceled'].value_counts()
```

Bookings that are canceled have `arrival_date` later than the date when the reservation status changes.

Some of the bookings which were not canceled have dates wrong, let's correct the dates based on the `days_diff` and `total_stay_days`.

```python
# add the number of total stay days to the `arrival_date` to get the corrected `reservation_status_date`, only apply to bookings with 'check-out' status
df['reservation_status_date'] = df.apply(lambda x: x.arrival_date+pd.Timedelta(days=x.total_stay_days) if x.reservation_status=='Check-Out' else x.reservation_status_date,axis=1)
```

```python
# drop the columns were used for this cleaning process
df = df.drop(columns=['arrival_date','days_diff','total_stay_days'])

# check
df.sample(3)
```

`reservation_status_date` contains multiple information in one column, should be separated into individual columns with just one information.

```python
# separate `reservation_status_date` into year, month, day
df['reservation_status_date_year'] = df['reservation_status_date'].dt.year
df['reservation_status_date_month'] = df['reservation_status_date'].dt.month
df['reservation_status_date_day'] = df['reservation_status_date'].dt.day

# drop original column
df = df.drop(columns='reservation_status_date')

# check
df.info()
```

```python
df['country'].unique()
```

Note that both 'CN' and 'CHN' mean China in the ISO 31553:2013 documentation. They should be uniformed, since all other country codes are in 3-letter format, I am keeping 'CHN'.

```python
# change 'CN' to 'CHN'
df['country'] = df['country'].replace('CN','CHN')

# check
df['country'].unique()
```

## Repeated Guests

According to the documentations of the dataset, a guest can only be categorized as repeated guest, if they have booked and stayed in the same hotel on at least one occasion. Let's check if there is any defective data regarding this.

```python
df[(df['is_repeated_guest']==1) & (df['previous_bookings_not_canceled']==0)]
```

```python
# get the guests who did not stay in the hotel but still categorized as repeated guests
defective_data = df[(df['is_repeated_guest']==1) & (df['previous_bookings_not_canceled']==0)]

print(f"There are {len(defective_data)} defective rows.")
```

Let's correct these data by reassigning the correct label of `is_repeated_guest` column.

```python
# reassign labels
df.loc[(df['is_repeated_guest'] == 1) & (df['previous_bookings_not_canceled'] == 0), 'is_repeated_guest'] = 0

# check
print(f"There are {len(df[(df['is_repeated_guest']==1) & (df['previous_bookings_not_canceled']==0)])} defective rows.")
```
